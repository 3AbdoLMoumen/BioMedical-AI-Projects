```python
import cv2
import numpy as np
from tensorflow.keras.layers import * 
from tensorflow.keras.models import * 
from tensorflow.keras.optimizers import * 
from tensorflow.keras.losses import * 
from tensorflow.keras.metrics import * 
import matplotlib.pyplot as plt
import os
from tqdm import tqdm
from sklearn.model_selection import train_test_split
import math
import warnings
warnings.filterwarnings("ignore")
```

```python
TRAIN_PATH = "/kaggle/input/brisc2025/brisc2025/segmentation_task/train"
TEST_PATH = "/kaggle/input/brisc2025/brisc2025/segmentation_task/test"
```

```python
def load_data(fname, path, size=(224, 224)):
    img_path = os.path.join(path, "images/", fname.replace("png", "jpg"))
    mask_path = os.path.join(path, "masks/", fname.replace("jpg", "png"))
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)/255
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
    mask[mask>0]=1
    img = cv2.resize(img, size)
    mask = cv2.resize(mask, size)
    img = np.expand_dims(img, -1).astype("float32")
    mask = np.expand_dims(mask, -1).astype("float32")

    return img, mask
```

```python
train_names = os.listdir(os.path.join(TRAIN_PATH, "images"))
test_names = os.listdir(os.path.join(TEST_PATH, "images"))

train_images, train_masks = [], []
test_images, test_masks = [], []

print("LOADING TRAIN DATA")
for fname in tqdm(train_names):
    img, mask = load_data(fname, TRAIN_PATH)
    train_images.append(img)
    train_masks.append(mask)
print("LOADING TEST DATA")
for fname in tqdm(test_names):
    img, mask = load_data(fname, TEST_PATH)
    test_images.append(img)
    test_masks.append(mask)

train_images, test_images, train_masks, test_masks = np.array(train_images), np.array(test_images), np.array(train_masks), np.array(test_masks)
train_images, val_images, train_masks, val_masks = train_test_split(train_images, train_masks, test_size=0.2, shuffle=True)
print(f"NUMBER OF TRAIN SAMPLES : {train_images.shape[0]}\nNUMBER OF VALIDATION SAMPLES : {val_images.shape[0]}\nNUMBER OF TEST SAMPLES : {test_images.shape[0]}")
```

```python
def build_unet(input_shape=(224,224,1), num_classes=1):
    inputs = Input(input_shape)

    c1 = Conv2D(64, (3,3), activation='relu', padding='same')(inputs)
    c1 = Conv2D(64, (3,3), activation='relu', padding='same')(c1)
    p1 = MaxPooling2D((2,2))(c1)

    c2 = Conv2D(128, (3,3), activation='relu', padding='same')(p1)
    c2 = Conv2D(128, (3,3), activation='relu', padding='same')(c2)
    p2 = MaxPooling2D((2,2))(c2)

    c3 = Conv2D(256, (3,3), activation='relu', padding='same')(p2)
    c3 = Conv2D(256, (3,3), activation='relu', padding='same')(c3)
    p3 = MaxPooling2D((2,2))(c3)

    c4 = Conv2D(512, (3,3), activation='relu', padding='same')(p3)
    c4 = Conv2D(512, (3,3), activation='relu', padding='same')(c4)
    p4 = MaxPooling2D((2,2))(c4)

    c5 = Conv2D(1024, (3,3), activation='relu', padding='same')(p4)
    c5 = Conv2D(1024, (3,3), activation='relu', padding='same')(c5)

    u6 = Conv2DTranspose(512, (2,2), strides=(2,2), padding='same')(c5)
    u6 = concatenate([u6, c4])
    c6 = Conv2D(512, (3,3), activation='relu', padding='same')(u6)
    c6 = Conv2D(512, (3,3), activation='relu', padding='same')(c6)

    u7 = Conv2DTranspose(256, (2,2), strides=(2,2), padding='same')(c6)
    u7 = concatenate([u7, c3])
    c7 = Conv2D(256, (3,3), activation='relu', padding='same')(u7)
    c7 = Conv2D(256, (3,3), activation='relu', padding='same')(c7)

    u8 = Conv2DTranspose(128, (2,2), strides=(2,2), padding='same')(c7)
    u8 = concatenate([u8, c2])
    c8 = Conv2D(128, (3,3), activation='relu', padding='same')(u8)
    c8 = Conv2D(128, (3,3), activation='relu', padding='same')(c8)

    u9 = Conv2DTranspose(64, (2,2), strides=(2,2), padding='same')(c8)
    u9 = concatenate([u9, c1])
    c9 = Conv2D(64, (3,3), activation='relu', padding='same')(u9)
    c9 = Conv2D(64, (3,3), activation='relu', padding='same')(c9)

    outputs = Conv2D(num_classes, (1,1), activation='sigmoid')(c9)

    model = Model(inputs=[inputs], outputs=[outputs])
    return model
```

```python
model = build_unet()
model.summary()
```

```python
loss = BinaryFocalCrossentropy(
    apply_class_balancing=True,
    alpha=0.5,
)
```

```python
model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss=loss,
    metrics=[BinaryIoU(target_class_ids=[0,1], threshold=0.5, name="iou"), "accuracy"]
)
```

```python
history = model.fit(train_images, train_masks, epochs=20, validation_data=(val_images, val_masks), batch_size=32)
```

```python
hist = history.history

metric_keys = [k for k in hist.keys()][:3]

n_metrics = len(metric_keys)
fig, axes = plt.subplots(1, n_metrics, figsize=(5*n_metrics, 4))

if n_metrics == 1:
    axes = [axes] 
for ax, key in zip(axes, metric_keys):
    ax.plot(hist[key], label=f"Train {key}", color="green", alpha=0.6)
    val_key = f"val_{key}"
    if val_key in hist:
        ax.plot(hist[val_key], label=f"Val {key}", color="red", alpha=0.6)
    ax.set_title(key)
    ax.set_xlabel("Epochs")
    ax.set_ylabel(key)
    ax.legend()
    ax.grid()
plt.tight_layout()
plt.show()
```

```python
model.evaluate(test_images, test_masks)
```

```python
model_preds = model.predict(test_images)  
binary_preds = (model_preds > 0.5).astype(np.uint8).squeeze(-1)  

gt_masks = test_masks
if gt_masks.ndim == 4 and gt_masks.shape[-1] == 1:
    gt_masks = np.squeeze(gt_masks, axis=-1)
gt_masks = gt_masks.astype(np.uint8)
```

```python
n_samples = 100  
pred_display_masks = np.zeros((binary_preds.shape[0], binary_preds.shape[1], binary_preds.shape[2], 3), dtype=np.uint8)
true_display_masks = np.zeros((gt_masks.shape[0], gt_masks.shape[1], gt_masks.shape[2], 3), dtype=np.uint8)

for i in range(binary_preds.shape[0]):
    pred_display_masks[i][binary_preds[i].astype(bool)] = (255, 0, 0) 
    true_display_masks[i][gt_masks[i].astype(bool)] = (0, 255, 0)       

cols = 4  
rows = math.ceil((n_samples * 2) / cols)
plt.figure(figsize=(cols*5, rows*4))

for i in range(n_samples):
    # True overlay
    plt.subplot(rows, cols, i*2 + 1)
    plt.imshow(test_images[i], cmap='gray')
    plt.imshow(true_display_masks[i], alpha=0.5)
    plt.title("True")
    plt.axis('off')

    # Pred overlay
    plt.subplot(rows, cols, i*2 + 2)
    plt.imshow(test_images[i], cmap='gray')
    plt.imshow(pred_display_masks[i], alpha=0.5)
    plt.title("Pred")
    plt.axis('off')
    plt.savefig(f'preds/overlay{i}.png', dpi=200) 
plt.tight_layout()
plt.show()
```
